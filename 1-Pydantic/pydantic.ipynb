```python
from pydantic import BaseModel
```

```python
from dataclasses import dataclass
@dataclass
class Person():
    name: str
    age: int 
    city: str

person = Person(name="Ankit", age=22, city="Pune")
print(person)
```

```python
class Person(BaseModel):
    name: str
    age: int 
    city: str

person = Person(name="Ankit", age=22, city="Pune")
print(person)
```
